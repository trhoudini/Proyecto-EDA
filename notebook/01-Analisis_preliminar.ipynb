### Columnas del dataset

●	``age``: La edad del cliente.

●	``job``: La ocupación o profesión del cliente.

●	``marital``: El estado civil del cliente.

●	``education``: El nivel educativo del cliente.

●	``default``: Indica si el cliente tiene algún historial de incumplimiento de pagos (1: Sí, 0: No).

●	``housing``: Indica si el cliente tiene un préstamo hipotecario (1: Sí, 0: No).

●	``loan``: Indica si el cliente tiene algún otro tipo de préstamo (1: Sí, 0: No).

●	``contact``: El método de contacto utilizado para comunicarse con el cliente.

●	``duration``: La duración en segundos de la última interacción con el cliente.

●	``campaign``: El número de contactos realizados durante esta campaña para este cliente.

●	``pdays``: Número de días que han pasado desde la última vez que se contactó con el cliente durante esta campaña.

●	``previous``: Número de veces que se ha contactado con el cliente antes de esta campaña.

●	``poutcome``: Resultado de la campaña de marketing anterior.

●	``emp.var.rate``: La tasa de variación del empleo.

●	``cons.price.idx``: El índice de precios al consumidor.

●	``cons.conf.idx``: El índice de confianza del consumidor.

●	``euribor3m``: La tasa de interés de referencia a tres meses.

●	``nr.employed``: El número de empleados.

●	``y``: Indica si el cliente ha suscrito un producto o servicio (Sí/No).

●	``date``: La fecha en la que se realizó la interacción con el cliente.

●	``contact_month``: Mes en el que se realizó la interacción con el cliente durante la campaña de marketing.

●	``contact_year``: Año en el que se realizó la interacción con el cliente durante la campaña de marketing.

●	``id_``: Un identificador único para cada registro en el dataset.

### Observaciones rápidas:

* `age`: El valor mínimo es 17, lo cual es razonable ya que algunas personas pueden abrir cuentas bancarias siendo mayores de edad. El máximo es 98, lo cual también es posible aunque inusual. La distribución parece razonable, con una media de ~40 años.

* `default`: Tiene solo valores 0 y 1, lo cual es correcto para una variable binaria. Sin embargo, la media es 0.00, lo que sugiere que casi ningún cliente tiene historial de impago. Podría tratarse de una variable poco informativa para el modelo si casi todos los valores son 0.

* `housing`: También es una variable binaria. La media es 0.54, lo que indica que el 54% de los clientes tienen hipoteca. Parece una distribución bastante equilibrada.

* `loan`: Otra variable binaria. Su media es 0.16, lo que implica que solo el 16% de los clientes tiene préstamos distintos a hipotecas. La variable está ligeramente desbalanceada, pero sigue siendo útil.

* `duration`: Muestra una gran dispersión (std similar a la media). El máximo es 4918 segundos (~82 minutos), lo cual podría ser un error o una llamada extraordinariamente larga. La mediana (179 s) es mucho menor que la media (257 s), lo que sugiere una distribución sesgada a la derecha.

* `campaign`: Tiene una media de 2.57 contactos por cliente. Sin embargo, hay valores máximos tan altos como 56, lo que indica posibles casos extremos que podrían distorsionar los modelos. Una revisión de outliers sería recomendable.

* `pdays`: La mediana y percentiles 25 y 75 están todos en 999, lo cual sugiere que este valor se está usando como marcador para "nunca contactado anteriormente". Esto debería tratarse como un valor especial y no como un número continuo en sí.

* `previous`: La mayoría de los valores son 0, con una media muy baja (0.17) y una mediana de 0. Esto indica que la mayoría de los clientes no han sido contactados previamente.

* `emp.var.rate`: Varía entre -3.4 y 1.4, lo cual es razonable. Es una variable económica externa, por lo tanto tiene sentido que sea más estática y esté centrada en torno a valores bajos.

### Observaciones rápidas:

* `job`: Hay 11 tipos distintos de ocupación. Los trabajos más comunes son:
    * `admin.` (24.9%)
    * `blue-collar` (22.1%)
    * `technician` (16.1%)
  Estos tres representan más del 60% de los registros, lo que sugiere una fuerte concentración en ciertos perfiles laborales.

* `marital`: Existen 3 estados civiles principales (`married`, `single`, `divorced`), con un fuerte predominio de casados (~60%). También hay 85 valores nulos (`NaN`), que conviene imputar o analizar.

* `education`: Se observan 7 niveles distintos, aunque hay redundancia en la codificación de “básico” (`basic.4y`, `basic.6y`, `basic.9y`). El 4.2% de los datos son nulos (`NaN`).`university degree` y `high school` son los niveles más frecuentes.

* `contact`: Solo hay 2 formas de contacto: `cellular` y `telephone`, con clara preferencia por el contacto móvil (~64%). Es una variable binaria clara.

* `poutcome`: Se refiere al resultado de campañas anteriores. La mayoría (~86%) tiene el valor `nonexistent`, lo que indica que no fueron contactados previamente o no hubo campaña anterior. Esto podría ser un buen predictor binario: ¿fue contactado antes o no?

* `cons.price.idx` y `cons.conf.idx`: Aunque técnicamente son numéricas, las estás tratando como categóricas probablemente porque tienen pocos valores únicos. Ambas tienen patrones de repetición y algunos `NaN`, lo que podría deberse a fechas específicas o ciclos económicos. Deberían analizarse en conjunto con `date`.

* `euribor3m`: Aunque tiene 309 valores únicos (numéricos), hay una cantidad significativa de nulos (~21%). Es un indicador económico muy útil para modelado, pero conviene imputarlo si se va a usar.

* `nr.employed`: Tiene solo 11 valores distintos, lo que sugiere que también es una variable de contexto económico general. Podría tratarse como variable categórica ordinal o numérica.

* `y`: Variable objetivo del modelo (target). Está desbalanceada: solo ~11% de los clientes han dicho `yes`, lo que sugiere la necesidad de técnicas para manejar ese desbalanceo.

* `date`: Tiene 1.825 fechas distintas, aunque algunas fechas como `28-febrero-2018` aparecen con mayor frecuencia. Hay 248 nulos. Esta columna puede descomponerse en `día`, `mes` y `año` para análisis temporal más detallado.

* `id_`: Todos los valores son únicos, por lo tanto no aporta valor predictivo directo. Se puede conservar como identificador si se desea rastrear registros.


### Countplot

```python
df = pd.read_csv('../data/raw/bank-additional.csv')
```

```python
columnas_cat = df.select_dtypes(include=['category', 'object']).columns
```

```python
for col in columnas_cat:

    if df[col].nunique() > 200:
        print(f"Columna {col} tiene demasiadas categorías ({df[col].nunique()}\n\n")
        continue

    print(f"Distribución de la columna {col}")
    plt.figure(figsize=(df[col].nunique() * 2, 3))
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.show()
```
